<h4> Définir une fonction </h4>

la fonction f(x) : f ⟼ R^(2), cette fonctiom calcule le périmètre d'un cercle 

```python
list_test = []

def f(R):
    return R*3

#Representation graphique 
for i in range(100):
    list_test.append(f(R=i))

pd.DataFrame(list_test).plot(kind="line", figsize=(6,6))

print(list_test)
 
```

```python
def f(R):
    if R <= -5:
        valeur = R**2
    elif -5 < R <= 12 :
        valeur = 3 * R
    elif R > 12:
        valeur = R**(0.5)
    return valeur
    

#Representation graphique 
for i in range(100):
    list_test.append(f(R=i))

pd.DataFrame(list_test).plot(kind="line", figsize=(6,6))

print(list_test)
```
