```python
#importer les librairies 
import yfinance as yf 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
ticker = ["AAPL"]
df = yf.download(ticker)["Adj Close"]
```

```python
rdt = np.log(df/df.shift(1))
rdt.dropna(inplace = True)
mu , sigma = rdt.mean() , rdt.std()
```

```python
initial_price = df.iloc[-1]
```

```python
for i in range(100):
    rdt_sim = np.random.normal(mu,sigma,252)
    price_sim = initial_price * (1 + rdt_sim).cumprod()
    plt.plot(price_sim)
    plt.axhline(initial_price, c = "red")
```
