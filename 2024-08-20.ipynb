Importer la liste des entreprises du S&P 500

```python
#importer les librairies 
import yfinance as yf
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
sp500url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
data = pd.read_html(sp500url)
```

```python
tickers = data[0]["Symbol"].tolist()
tickers
```

```python
for i in range(len(tickers)):
    if tickers[i] == "BRK.B":
        tickers[i] = "BRK-B"
    elif tickers[i] =="BF.B":
        tickers[i] = "BF-B"
```

```python
#definir l'environnement
start_date = "2023-01-01"
end_date = "2024-01-01"

price = yf.download(tickers,start=start_date,end=end_date, interval="1mo")["Adj Close"]

price

```

```python
price.loc[:, price.isna().any()]
```

```python
price.dropna()
```

```python
rendement = np.log(price/price.shift(1))
```

```python
rendement
```

```python
rdt_moyen = rendement.mean()
std_rendement = rendement.std()
print("------")
print(rdt_moyen.argmax())
print(rdt_moyen.iloc[rdt_moyen.argmax()])
print(std_rendement.iloc[rdt_moyen.argmax()])
```

```python
print("--------")
print(rdt_moyen.argmin())
print(rdt_moyen.iloc[std_rendement.argmin()])
print( std_rendement.iloc[std_rendement.argmin()])
```

```python
Ascending = rdt_moyen.sort_values(ascending=False)
```

```python
Top_5 = pd.DataFrame(Ascending.head())
Top_5
```

```python
Top_5_liste = ["SMCI", "NVDA","CRWD","META","PLTR"]
price_top_5 = yf.download(Top_5_liste,start="2024-01-01",end="2024-12-01", interval="1mo")["Adj Close"]
rdt_top_5 = np.log(price_top_5/price_top_5.shift(1))
mean_top_5 = np.array(rdt_top_5.mean())
var_top_5 = np.array(rdt_top_5.var())
print(mean_top_5, var_top_5)
```

```python
plt.scatter(var_top_5,mean_top_5)
plt.xlabel("Variance")
plt.ylabel("Rendement")
```

```python
price_top_5.plot()
```

```python
CovMat = rdt_top_5.cov()
inv_CovMat = np.linalg.inv(CovMat)
VU = np.ones((5,1))
```

```python
GMV = (inv_CovMat@VU)/(VU.T@inv_CovMat@VU)
Er_GMV = GMV.T@mean_top_5
Std_GMV = np.sqrt(np.dot(np.dot(GMV.T,CovMat),GMV))
```

```python
Sharpe = (inv_CovMat@mean_top_5.T)/(VU.T@inv_CovMat@mean_top_5)
Er_Sharpe = Sharpe.T@mean_top_5
Std_Sharpe = np.sqrt(np.dot(np.dot(Sharpe.T,CovMat),Sharpe))
```

```python
Cov_GMV_Sharpe = np.array(GMV.T@CovMat@Sharpe)

w_GMV = np.linspace(-2,3,100)
w_Sharpe = 1 - w_GMV

Er_fr_eff = (w_GMV * Er_GMV + w_Sharpe * Er_Sharpe)
Std_fr_eff = np.sqrt(((w_GMV*Std_GMV)**2+(w_Sharpe*Std_Sharpe)**2+2*w_GMV*w_Sharpe*Cov_GMV_Sharpe))
```

```python
plt.scatter(Std_fr_eff,Er_fr_eff,)
plt.scatter(Std_GMV, Er_GMV, color = "red")
plt.scatter(Std_Sharpe, Er_Sharpe, color = "red")
plt.show()
```
